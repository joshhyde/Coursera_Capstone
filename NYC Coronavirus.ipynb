```python
import pandas as pd 

url = 'https://github.com/nytimes/covid-19-data/raw/master/us-counties.csv'
df = pd.read_csv(url)

df.head()
```

```python
nyc = df[df['county'] == 'New York City']

nyc
```

```python
nyc.drop(columns=['fips'], inplace=True)
nyc.dtypes
```

```python
import matplotlib.pyplot as plt

nyc_deaths = nyc['deaths']
nyc_cases = nyc['cases']

#fig = plt.figure()
#ax = plt.axes()
plt.rcParams["figure.figsize"] = (15,10)
nyc[["date", "deaths"]].plot(x="date")
plt.xticks(rotation=45)


plt.show()
```

```python
nyc[["date", "cases"]].plot(x="date")

plt.show()
```
